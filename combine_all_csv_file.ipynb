```python
import os
import pandas as pd

def concatenate_csv_files(root_dir):
    # Initialize an empty DataFrame to hold all the concatenated data
    combined_df = pd.DataFrame()

    # Walk through the root directory
    for subdir, _, files in os.walk(root_dir):
        # Check if 'all_data_46_46.csv' exists in the current directory
        if 'all_data_45_46.csv' in files:
            file_path = os.path.join(subdir, 'all_data_45_46.csv')
            print(f"Processing file: {file_path}")

            # Read the CSV file
            df = pd.read_csv(file_path)

            # Concatenate the current dataframe with the combined dataframe
            combined_df = pd.concat([combined_df, df], axis=0, ignore_index=True)

    return combined_df

# Specify the root directory containing the subfolders
root_directory = 'D:/20231112_LBNL_COLTRIMS/202308_ALS_APR_2023_HCOOH Analysis/DATA/HCOOH/20240123_HCOOH_dimer/'

# Get the concatenated DataFrame
result_df = concatenate_csv_files(root_directory)

print(result_df)

output_file = os.path.join(root_directory, "combined_all_data_45_46.csv")

# Optionally, save the combined data to a new CSV file
result_df.to_csv(output_file, index=False)
```
